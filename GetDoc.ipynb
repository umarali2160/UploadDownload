```python
def deletecol(dataframe):
    new_df = dataframe.drop(columns = ['Open','High','Low','Close','Volume'])
    return new_df
```

```python
import pandas as pd
import base64
from IPython.display import display, HTML
def create_download_link( df, title = "Download CSV file", filename = "data.csv"):  
    csv = df.to_csv()
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)
```

```python
import pandas as pd
from tkinter import *
from tkinter.ttk import *
from tkinter.filedialog import askopenfile 

# root = Tk()

# root.title('Explorer')
# root.geometry("500x500")
# root.config(background = "white")

# upload_button = Button(root, 
#     text="Upload File", 
#     command=lambda:UploadAction()
#     ).grid(row=0,column=1)


def UploadAction():
    file_path = askopenfile(mode='r', filetypes=[('Image Files', '*csv')])
    dataframe = pd.read_csv(file_path.name)  
    # print(dataframe.head())
    new_df = deletecol(dataframe)
    print(new_df.head())
    return new_df

# root.mainloop()
```

```python
def main():
    root = Tk()

    root.title('Explorer')
    root.geometry("500x500")
    root.config(background = "white")

    upload_button = Button(root, 
        text="Upload File", 
        command=lambda:UploadAction()
        ).grid(row=0,column=1)

    df = UploadAction()
    root.mainloop()
    return df
```

```python
if __name__ == "__main__":
    df = main()
create_download_link(df)
```
